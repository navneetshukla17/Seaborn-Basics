```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = sns.load_dataset('tips')
data.head()
```

# Bar Plot

```python
sns.catplot(data=data, x='smoker', y='total_bill', kind='bar')
```

```python
sns.catplot(data=data, x='smoker', y='total_bill', kind='bar', hue='sex')
```

### changing estimators 

```python
sns.catplot(data=data, x='smoker', y='total_bill', kind='bar', hue='sex', estimator=np.median)
```

```python
sns.catplot(data=data, x='smoker', y='total_bill', kind='bar', hue='sex', estimator=np.var) # for variance
```

```python
sns.catplot(data=data, x='smoker', y='total_bill', kind='bar', hue='sex', estimator=np.std) #for standard deviation
```

# Count Plot
- Count plots works as histograms work for numerical data, It use to count categorical data like histograms does with numerical where it also counts the numerical data

```python
sns.catplot(data=data, x='sex', kind='count') # total no. of males and females
```

```python
sns.catplot(data=data, x='sex', kind='count', hue='smoker') # no. of smokers from male and female
```
