# Pairplots

Pairplots are used to plot multiple pairwise bivariate distribution in a dataset. This creates a matrix of axes and shows the relationship for each pair of columns in a DataFrame. By defualt it also draws the univariate distribution of each variable on the diagonal axes.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
plt.style.use('bmh')
```

```python
data = sns.load_dataset('iris')
data.head()
```

```python
sns.pairplot(data, hue='species')
```
