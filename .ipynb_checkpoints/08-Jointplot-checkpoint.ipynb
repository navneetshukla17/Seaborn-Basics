# Jointplot
 
Joinplot creates a multi-panel figure that shows both the bivariate (or joint) relationship between two variables along with the univariate (or marginal) distribution of each on separate axes.

We can use jointplots in the following 4 manners:

1. Scatter plots
2. Hexbin plots
3. Kde plots
4. Reg plots

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = sns.load_dataset('tips')
data.head()
```

```python
# basic joint plot
sns.jointplot(data=data, x='total_bill', y='tip')
```

```python
# Hexbin plot
sns.jointplot(data=data, x='total_bill', y='tip', kind='hex')
```

```python
sns.jointplot(data=data, x='total_bill', y='tip', kind='kde')
```

```python
sns.jointplot(data=data, x='total_bill', y='tip', kind='reg')
```
