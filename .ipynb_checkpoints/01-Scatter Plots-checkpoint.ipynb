# Scatter Plots

The scatter plot is a mainstay of statistical visualization. It depicts the joint distribution of two sometimes three variables using a cloud of points, where each point represents an observation in the dataset. This depiction allows the eye to infer a substantial amount of information about wheter there is any meaningful relationship between them

Although, scatter plot can be plotted for both numerical as well as categorical data, it's mostly used to depict relationship between numerical data.

There are 2 ways to draw a scatterplot in seaborn

### 1. relplot func()[Figure Level Function] 
### 2. scatterplot func() [Axes Level Function]

As we will see, these visualizations can be quite helpful because they use simply understood represantations of data even on complex datasets by mapping up to three additional variables using the semantics of `hue`, `size` and `style`

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = sns.load_dataset('tips')
data.head()
```

### Plot a scatter plot between `total_bill` and `tip`

```python
sns.relplot(data=data, x='total_bill', y='tip')
```

### The hue parameter

```python
sns.relplot(data=data, x='total_bill', y='tip', hue='smoker')
```

### The style parameter

```python
sns.relplot(data=data, x='total_bill', y='tip', hue='smoker', style='sex')
```

### The size parameter

```python
sns.relplot(data=data, x='total_bill', y='tip', hue='smoker', style='sex', size='size')
```

### Normalize the  variables

```python
sns.relplot(data=data, x='total_bill', y='tip', hue='smoker', style='sex', size='size', sizes=(15, 200))
```

### scatter plot

```python
sns.scatterplot(data=data, x='total_bill', y='tip', hue='smoker', style='sex', size='size')
```
