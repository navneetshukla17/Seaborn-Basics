# Distplots / Distribution Plots

Distplots are used to do univariate analysis on a continous/numerical data. Distplots are combination of 3 things in one graph:
1. Histogram
2. Kernel Density Estimation (PDF)
3. Rugplot

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('titanic')
data.head()
```

```python
# data[data['age'].isnull()][['age']] # total 177 missing values in `age` column 
```

```python
data['age'].fillna(data['age'].mean(), inplace=True)
```

## Plotting a Distplot

```python
sns.distplot(data['age'])
```

## Plotting only the Histogram

```python
sns.distplot(data['age'], kde=False)
```

## Plotting only the KDE

```python
sns.distplot(data['age'], hist=False)
```

## Plotting only the rug plot

```python
sns.distplot(data['age'], rug=True, hist=False, kde=False)
```

## How to use Distplot

```python
sns.distplot(data[data['survived'] == 1]['age']) # displot on survived passengers
```

### Plotting a Distplot on Comparision between `Passengeres who survived` Vs `Passengers who died`

```python
sns.distplot(data[data['survived'] == 1]['age'], label='Survived')
sns.distplot(data[data['survived'] == 0]['age'], label='Died')
plt.legend(loc='upper left')
```
