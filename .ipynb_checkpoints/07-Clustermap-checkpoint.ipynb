# Clustermap

It is basicly a heatmap but it is clustered, it is herarchly clustered it clustereds rows and columns together on the basis of similarity, So the similar rows and columns will come together

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('flights')
data.head()
```

```python
x = data.pivot_table(index='year', columns='month', values='passengers', aggfunc='sum')
```

```python
sns.clustermap(x)
```

```python
sns.clustermap(x, metric='correlation')
```

```python
sns.clustermap(x, metric='correlation', z_score=0)
```

```python
sns.clustermap(x, linewidths=0.5, metric='correlation', z_score=0, annot=True)
```

```python
# row clustering
sns.clustermap(x, linewidths=0.5, metric='correlation', z_score=0, annot=True, col_cluster=False)
```

```python
# column cluster
sns.clustermap(x, linewidth=0.5, metric='correlation', z_score=0, annot=True, row_cluster=False)
```
