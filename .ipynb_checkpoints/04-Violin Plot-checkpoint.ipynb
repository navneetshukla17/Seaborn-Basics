# Violin Plots

- Violin Plots are a method of plotting numeric data and can be considerd as combination of the box plot with kde [kernel density equation] plot

- The unquestionable advantages of the violin plot over box plot is that aside from showing the abovementioned statistics it also shows the entire distribution of the data.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = sns.load_dataset('tips')
data.head()
```

```python
sns.catplot(data=data, x='day', y='total_bill', kind='violin')
```

```python
sns.catplot(data=data, x='day', y='total_bill', kind='violin', hue='sex')
```

```python
# after applying split
sns.catplot(data=data, x='day', y='total_bill', kind='violin', hue='sex', split=True)
```
