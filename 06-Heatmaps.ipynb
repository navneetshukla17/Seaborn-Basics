# Heatmap

A heatmap is a graphical represantation of data where individual values contained in a matrix are represented as colors.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('flights')
data.head(5)
```

```python
x = data.pivot_table(index='year', columns='month', values='passengers', aggfunc='sum')
```

```python
sns.heatmap(x)
```

```python
# Removing color bar
sns.heatmap(x, cbar=False)
```

```python
# applying linewidths
sns.heatmap(x, linewidths=0.1)
```

```python
# add anotations 
sns.heatmap(x, annot=True, fmt='d', linewidths=0.5) # total no. of passengers in every month of every year
```

```python
# changing color tones
sns.heatmap(x, linewidth=0.5, annot=True, fmt='d', cmap='viridis')
```

```python
sns.heatmap(x, linewidths=0.5, annot=True, fmt='d', cmap='summer')
```

```python
sns.heatmap(x, linewidths=0.5, annot=True, fmt='d', cmap='winter')
```
